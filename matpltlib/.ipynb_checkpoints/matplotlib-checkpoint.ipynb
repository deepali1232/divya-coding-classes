```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
                        #add titles and labels
plt.title("line plot")
plt.xlabel('xlabel')
plt.ylabel('y label')
plt.show()
```

```python
#changing line aesthetics
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.show()
```

```python
#add two lines in the same plot
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=3)
plt.title('line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

```python
#adding sub plot
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y1,color='r',linestyle=':',linewidth=2)
plt.show()
```

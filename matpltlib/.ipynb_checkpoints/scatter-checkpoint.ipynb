```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=[1,2,4,5,6,7,8]
a=[1,3,4,5,6,7,7]
plt.scatter(x,a,marker='*',c='g',s=100)
plt.show()
```

```python
x=[1,2,4,5,6,7,8]
a=[1,3,4,5,6,7,7]
b=[1,2,3,4,5,6,7]
plt.scatter(x,a,marker='*',c='r',s=100)
plt.scatter(x,b,marker='*',c='g',s=200)
plt.show()
```

```python
x=[1,2,4,5,6,7,8]
a=[1,3,4,5,6,7,7]
b=[1,2,3,4,5,6,7]
plt.subplot(1,2,1)
plt.scatter(x,b,marker='*',c='g',s=200)
plt.subplot(1,2,2)
plt.scatter(x,b,marker='*',c='r',s=400)
plt.show()
```
